```python
# face_recognition from image
import os
import cv2
from keras.models import load_model
import numpy as np

def predict(img_path, h5_path):
    model = load_model(h5_path)
    img = cv2.imread(img_path)
    img = cv2.resize(img, (128, 128))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = img.reshape((1, 1, 128, 128))
    img = img.astype('float32')
    img = img/255.0
    prob = model.predict(img)
    label = np.argmax(model.predict(img))
    return prob, label

print(predict("E:/test/n000001/0.jpg", "D:/桌面/FaceFinal.h5"))
```

```python
# face_recognition from camera
from keras.models import load_model

def recognition(h5_path):
    model = load_model(h5_path)
    color = (0, 255, 0)
    cap = cv2.VideoCapture(0)
    while True:
        # read one frame of the video
        ret, frame = cap.read()  

        if ret is True:

            # make the image gray
            frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        else:
            continue

        cascade = cv2.CascadeClassifier("D:\Anaconda\envs\py36\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml")

        # get the area of face
        faceRects = cascade.detectMultiScale(frame_gray, scaleFactor=1.2, minNeighbors=3, minSize=(32, 32))
        #print(faceRects)
        if len(faceRects) > 0:
            for faceRect in faceRects:
                x, y, w, h = faceRect

                # fit into the model to see who this is 
                img = frame[y - 10: y + h + 10, x - 10: x + w + 10]
                img = cv2.resize(img, (128, 128))
                img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                img = img.reshape((1, 1, 128, 128))
                img = img.astype('float32')
                img = img/255.0
                prob = max(model.predict(img)[0])
                label = str(np.argmax(model.predict(img)))

                # print('name_number:', name_number)
                cv2.rectangle(frame, (x - 10, y - 10), (x + w + 10, y + h + 10), color, thickness=2)
                print(str(prob)+"   "+str(label))

                # show who this is 
                #cv2.putText(frame, name, (x + 30, y + 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 255), 2)
                if prob > 0.99:
                    cv2.putText(frame, label, (x + 30, y + 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 255), 2)
                else:
                    cv2.putText(frame, 'unknow', (x + 30, y + 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 255), 2)

        cv2.imshow("face_recognition", frame)

        # wait for 10 ms to see if any keyboard input
        k = cv2.waitKey(10)
        # break with "q" as input
        if k & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
recognition("D:/桌面/model.h5")
```
