```python
import os
import cv2

def endwith(s,*endstring):
    resultArray = map(s.endswith,endstring)
    if True in resultArray:
        return True
    else:
        return False

def readAllImg(path,*format):
    try:

        s = os.listdir(path)
        resultArray = []
        fileName = os.path.basename(path)
        resultArray.append(fileName)

        for i in s:
            if endwith(i, format):
                document = os.path.join(path, i)
                img = cv2.imread(document)
                resultArray.append(img)

    except IOError:
        print ("Error, the file may not be found")

    else:
        #print ("successful")
        return resultArray
```
